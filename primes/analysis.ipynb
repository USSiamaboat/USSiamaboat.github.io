```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("results_50K.json")

bins = np.arange(10)
for col in df.columns:
    plt.hist(df[col], bins=bins, label=col, alpha=0.5)

plt.title("Finding primes up to 50,000 runtimes")
plt.xlabel("Time (ms)")

plt.legend()
plt.show()
```
